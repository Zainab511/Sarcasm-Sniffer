#**Extraction of Tweets using Scrapper API**

# Non-Sarcastic Tweets

```python
def scrape_tweets(api_key, query, num_tweets):
    payload = {
        'api_key': api_key,
        'query': query,
        'num': num_tweets,
    }
    response = requests.get('https://api.scraperapi.com/structured/twitter/search', params=payload)
    data = response.json()
    all_tweets = data['organic_results']
    return all_tweets

api_key = '5c27dfe6530b2c12ec97ef2eddac6c31'
num_tweets = '100'

queries = ['#happy', '#joy', '#positive', '#love', '#fun', '#excited', '#grateful', '#amazing', '#fantastic', '#cheerful', '#optimistic', '#kind', '#uplifting', '#friendly', '#vibrant', '#lively', '#goodvibes', '#peace', '#calm', '#content', '#gratitude', '#serene', '#motivated', '#success', '#laughter', '#delight', '#admirable', '#inspiration', '#blessed', '#positivity']
twitter_data = []

for query in queries:
    tweets = scrape_tweets(api_key, query, num_tweets)
    twitter_data.extend(tweets)

folder_path = '/content/drive/MyDrive/Realtime_Dataset/'

df = pd.DataFrame(twitter_data)

df.to_csv(folder_path + 'real_time_dataset1.csv', index=False)

print('File exported and saved in Google Drive at:', folder_path)

print(df)
```
